```python
import csv
```

```python
def is_spielberg(record):
    return record['director_last_name'] == 'Spielberg'

def print_film(record):
    print(record['movie_name'])
```

```python
with open('all_data.csv','rt') as csvfile:
    reader = csv.DictReader(csvfile)
    ff = filter(is_spielberg,reader)
    m = map(print_film,ff)
    
 
```
